#  **<h1 style='color:Indigo' align="center" > Data Processing </h1>**

```python
cols_yes__no_values = ['HeartDisease', 'Smoking', 'AlcoholDrinking', 'Stroke', 'DiffWalking', 'PhysicalActivity',
                       'Asthma', 'KidneyDisease', 'SkinCancer']
yes__no_values = {'No':0, 'Yes':1}
for i in range(0, len(cols_yes__no_values)):
    ourdata[cols_yes__no_values[i]] = ourdata[cols_yes__no_values[i]].replace(yes__no_values)
sex = {'Female':0, 'Male':1}
ourdata['Sex'] = ourdata['Sex'].replace(sex)

ageCategory= {'18-24':0, '25-29':1, '30-34':2, '35-39':3, '40-44':4, '45-49':5, '50-54':6, '55-59':7, '60-64':8,
             '65-69':9, '70-74':10, '75-79':11, '80 or older':12}
ourdata['AgeCategory'] = ourdata['AgeCategory'].replace(ageCategory)
genHealth = {'Poor':0, 'Fair':1, 'Good':2, 'Very good':3, 'Excellent':4}
ourdata['GenHealth'] = ourdata['GenHealth'].replace(genHealth)
diabetic = {'No':0, 'No, borderline diabetes':1, 'Yes (during pregnancy)':2, 'Yes':3}
ourdata['Diabetic'] = ourdata['Diabetic'].replace(diabetic)
ourdata.drop(columns=['Race'], inplace=True)
ourdata
```

```python
ourdata_majority_0 = ourdata[(ourdata['HeartDisease']==0)] 
ourdata_minority_1 = ourdata[(ourdata['HeartDisease']==1)] 

ourdata_minority_upsampled = resample(ourdata_minority_1, 
                                 replace=True,    
                                 n_samples= 292225, #292225
                                 random_state=42) 

ourdata_upsampled = pd.concat([ourdata_minority_upsampled, ourdata_majority_0])
```

```python
ourdata_upsampled
```

#  **<h1 style='color:Indigo' align="center" > Data Spliting </h1>**

```python
X = ourdata_upsampled.drop(columns='HeartDisease')
y = ourdata_upsampled['HeartDisease']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 15)
```

#  **<h1 style='color:Indigo' align="center" > Data Scaling </h1>**

```python

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
X_train = pd.DataFrame(X_train)
X_test = pd.DataFrame(X_test)
```
